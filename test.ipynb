```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
import matplotlib.pyplot as plt
from matplotlib.colors import to_hex  # ✅ ใช้อันนี้แทน folium.colors

# สมมติว่าคุณมี DataFrame ชื่อ df
# df = pd.read_csv('your_data.csv')

# สร้างสีสำหรับแต่ละ zone
zones = df_main['zone'].unique()
colors = plt.cm.tab10.colors  # สีสดใส 10 สี
zone_color_map = {zone: to_hex(colors[i % len(colors)]) for i, zone in enumerate(zones)}

# สร้างแผนที่
m = folium.Map(location=[14.037, 100.614], zoom_start=13)
marker_cluster = MarkerCluster().add_to(m)

# เพิ่มจุดแต่ละร้าน
for _, row in df_main.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['long']],
        radius=max(row['comment'] / 10, 3),  # ขนาดฟอง
        popup=f"<b>{row['name']}</b><br>Zone: {row['zone']}<br>Rating: {row['rating']}<br>Comment: {row['comment']}",
        color=None,
        fill=True,
        fill_color=zone_color_map.get(row['zone'], 'gray'),  # สีตาม zone
        fill_opacity=0.6
    ).add_to(marker_cluster)

m
```
