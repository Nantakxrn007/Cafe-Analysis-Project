```python
if not df_option.empty:
    print("\nกำลัง map option ดั้งเดิมใน df_option เข้ากับกลุ่มใหม่...")
    # คอลัมน์ที่เก็บ option text ดั้งเดิมใน df_option ของคุณคือ 'option'
    df_option['option_group'] = df_option['option'].map(option_to_group_mapping)

    print("\nตัวอย่าง df_option หลังจากเพิ่มคอลัมน์ 'option_group':")
    print(df_option.head())

    # ตรวจสอบ options ที่อาจจะ map ไม่ติด
    unmapped_options = df_option[df_option['option_group'].isnull()]['option'].unique()
    if len(unmapped_options) > 0:
        print(f"\nคำเตือน: พบ {len(unmapped_options)} options ใน df_option ที่ไม่ได้อยู่ใน option_to_group_mapping:")
        for opt in unmapped_options[:10]: # แสดง 10 รายการแรกที่ map ไม่ติด
            print(f"- '{opt}'")
        print("Options เหล่านี้จะทำให้เกิดค่า NaN ในคอลัมน์ 'option_group'")
        print("แนะนำให้เพิ่ม options เหล่านี้เข้าไปใน dictionary หรือเติมค่า NaN ด้วยกลุ่มเริ่มต้น")
        # ตัวอย่างการเติมค่า NaN ด้วย 'Unknown/Other'
        df_option['option_group'].fillna('Unknown/Other', inplace=True)
        print("ได้เติมค่า NaN ใน 'option_group' ด้วย 'Unknown/Other' แล้ว")
    else:
        print("\nOptions ทั้งหมดใน df_option ถูก map เข้ากลุ่มเรียบร้อยแล้ว")
else:
    print("\nไม่สามารถดำเนินการ map ได้ เนื่องจาก df_option ว่างเปล่า")
```

```python
if not df_option.empty and 'option_group' in df_option.columns:
    print("\nกำลังแปลงข้อมูลเป็น Wide Format (One-Hot Encoding option groups)...")

    # 1. เลือกคอลัมน์ที่จำเป็นและลบรายการซ้ำซ้อน
    # เพื่อให้แน่ใจว่าแต่ละร้านจะถูกนับเพียงครั้งเดียวสำหรับแต่ละ option group
    df_groups_per_cafe = df_option[['id_option_main', 'option_group']].drop_duplicates()

    # (เพิ่มเติม) จัดการกับกรณีที่ 'option_group' อาจจะเป็น NaN ถ้าคุณยังไม่ได้ fillna ในขั้นตอนก่อนหน้า
    # และไม่ต้องการให้ 'NaN' กลายเป็นชื่อคอลัมน์ feature
    df_groups_per_cafe = df_groups_per_cafe.dropna(subset=['option_group'])


    # 2. ทำ One-Hot Encoding (สร้าง Dummy Variables)
    # ใช้ pd.crosstab จะสร้างตารางที่นับความถี่ (ซึ่งจะเป็น 1 เพราะเรา drop_duplicates ไปแล้ว)
    df_option_group_features = pd.crosstab(df_groups_per_cafe['id_option_main'], df_groups_per_cafe['option_group'])

    # Reset index เพื่อให้ 'id_option_main' กลับมาเป็นคอลัมน์สำหรับ merge
    df_option_group_features = df_option_group_features.reset_index()

    print("\nตัวอย่าง df_option_group_features (ข้อมูล options ที่แปลงเป็น features แล้ว):")
    print(df_option_group_features.head())
    print(f"Shape of df_option_group_features: {df_option_group_features.shape}")
else:
    print("\nไม่สามารถสร้าง df_option_group_features ได้ เนื่องจาก df_option ว่างเปล่า หรือไม่มีคอลัมน์ 'option_group'")
    df_option_group_features = pd.DataFrame(columns=['id_option_main']) # สร้าง DataFrame ว่างไว้เผื่อกรณี merge
```

```python
if not df_main.empty and not df_option_group_features.empty and 'id_option_main' in df_main.columns and 'id_option_main' in df_option_group_features.columns:
    print("\nกำลังรวม df_main เข้ากับ df_option_group_features...")

    df_final_for_clustering = pd.merge(df_main, df_option_group_features, on='id_option_main', how='left')

    # หลังจาก merge, คอลัมน์ที่เป็น option group features อาจจะมีค่า NaN
    # สำหรับร้านใน df_main ที่ไม่มีข้อมูลใน df_option_group_features
    # เราจะเติมค่า NaN เหล่านี้ด้วย 0 (หมายถึงไม่มี feature กลุ่มนั้น)
    option_group_feature_columns = df_option_group_features.columns.drop('id_option_main', errors='ignore') # errors='ignore' เผื่อกรณีที่ไม่มีคอลัมน์นี้จริงๆ

    for col in option_group_feature_columns:
        if col not in df_final_for_clustering.columns:
             # กรณีที่ df_option_group_features อาจจะว่างเปล่าตั้งแต่แรก (เช่น ไม่มี options เลย)
             # หรือบางกลุ่มไม่มีร้านใดมีเลย ทำให้ไม่เกิดเป็นคอลัมน์
             df_final_for_clustering[col] = 0 # สร้างคอลัมน์ใหม่และเติม 0
        else:
             df_final_for_clustering[col] = df_final_for_clustering[col].fillna(0).astype(int)


    print("\nตัวอย่าง DataFrame สุดท้ายสำหรับเตรียมทำ Clustering (df_final_for_clustering):")
    print(df_final_for_clustering.head())
    print(f"Shape of df_final_for_clustering: {df_final_for_clustering.shape}")

    print("\nคอลัมน์ใน df_final_for_clustering:")
    print(df_final_for_clustering.columns.tolist())

    # แสดงตัวอย่าง features ใหม่ที่เพิ่มเข้ามา
    if len(option_group_feature_columns) > 0:
        sample_cols_to_show = ['id_option_main'] + [col for col in option_group_feature_columns if col in df_final_for_clustering.columns][:5]
        print("\nตัวอย่าง features ใหม่จาก option groups ใน DataFrame สุดท้าย:")
        print(df_final_for_clustering[sample_cols_to_show].head())
    else:
        print("\nไม่ได้มีการเพิ่มคอลัมน์ features จาก option groups (อาจมีปัญหาในขั้นตอนก่อนหน้า)")

elif df_main.empty:
    print("\nไม่สามารถทำการ merge ได้ เนื่องจาก df_main ว่างเปล่า")
else:
    print("\nไม่สามารถทำการ merge ได้ เนื่องจาก df_option_group_features ว่างเปล่า หรือไม่มีคอลัมน์ 'id_option_main' ที่ถูกต้อง")
    # ถ้า df_option_group_features ว่างเปล่าแต่ df_main มีข้อมูล, df_final_for_clustering จะเท่ากับ df_main
    df_final_for_clustering = df_main.copy()
    print("df_final_for_clustering จะเป็นสำเนาของ df_main เนื่องจากไม่มี features จาก options ให้ merge")
```

```python
import pandas as pd

# กำหนด Path ไปยังโฟลเดอร์ Final3 (ถ้าสคริปต์รันจาก marketing_final/marketing_007/)


# 1. โหลด DataFrame ตารางหลัก
file_mainlast =  'df_mainlast.csv'
try:
    df_mainlast = pd.read_csv(file_mainlast)
    print("โหลด df_mainlast สำเร็จ:")
    print(df_mainlast.head())
    print(f"คอลัมน์ใน df_mainlast: {df_mainlast.columns.tolist()}")
except FileNotFoundError:
    print(f"Error: ไม่พบไฟล์ {file_mainlast}")
    df_mainlast = pd.DataFrame() # สร้าง DataFrame ว่างถ้าหาไฟล์ไม่เจอ

# 2. โหลด DataFrame ตาราง Option (จาก df_option.csv)
#    เราจะตั้งชื่อ DataFrame ที่โหลดมานี้ว่า df_options_table เพื่อให้สื่อความหมาย
file_options =  df_option_grouped'
try:
    df_options_table = pd.read_csv(file_options)
    print("\nโหลด df_options_table (จาก df_option.csv) สำเร็จ:")
    print(df_options_table.head())
    print(f"คอลัมน์ใน df_options_table: {df_options_table.columns.tolist()}") # ควรจะมี 'id_option_main' และ 'option'
except FileNotFoundError:
    print(f"Error: ไม่พบไฟล์ {file_options}")
    df_options_table = pd.DataFrame() # สร้าง DataFrame ว่างถ้าหาไฟล์ไม่เจอ

# --- ตรวจสอบว่า DataFrame ทั้งสองมีข้อมูลและคอลัมน์ที่ถูกต้องก่อนทำการ Merge ---
if not df_mainlast.empty and not df_options_table.empty:
    # 3. ทำการ Merge ข้อมูล
    # ตรวจสอบว่าคอลัมน์ 'id_option_main' มีอยู่ในทั้งสอง DataFrame
    if 'id_option_main' in df_mainlast.columns and 'id_option_main' in df_options_table.columns:
        merged_df = pd.merge(
            df_mainlast,
            df_options_table,
            on='id_option_main', # ใช้คอลัมน์ 'id_option_main' ซึ่งมีชื่อเดียวกันในทั้งสองตาราง
            how='left'          # คงทุกแถวจาก df_mainlast และนำข้อมูลจาก df_options_table มาต่อท้าย
        )
        print("\nDataFrame หลังจากการ Merge (merged_df):")
        print(merged_df.head())

        # แสดงจำนวนแถวก่อนและหลัง merge เพื่อตรวจสอบ
        print(f"\nจำนวนแถวใน df_mainlast: {len(df_mainlast)}")
        print(f"จำนวนแถวใน df_options_table: {len(df_options_table)}")
        print(f"จำนวนแถวใน merged_df: {len(merged_df)}")

        # หากต้องการดูว่า ID ไหนใน df_mainlast ที่มีหลาย option
        # print("\nตัวอย่าง ID ที่มีหลาย options:")
        # print(merged_df[merged_df.duplicated(subset=['id_open_close_main'], keep=False)].head()) # หรือใช้ id_option_main ถ้าเป็นคีย์หลัก
    else:
        print("\nError: ไม่พบคอลัมน์ 'id_option_main' ใน DataFrame ใด DataFrame หนึ่ง หรือทั้งสอง")
        if 'id_option_main' not in df_mainlast.columns:
            print(f"คอลัมน์ 'id_option_main' ไม่ได้อยู่ใน df_mainlast. คอลัมน์ที่มีคือ: {df_mainlast.columns.tolist()}")
        if 'id_option_main' not in df_options_table.columns:
            print(f"คอลัมน์ 'id_option_main' ไม่ได้อยู่ใน df_options_table. คอลัมน์ที่มีคือ: {df_options_table.columns.tolist()}")

else:
    print("\nไม่สามารถทำการ merge ได้ เนื่องจาก DataFrame หนึ่งหรือทั้งสองตัวว่างเปล่าจากการโหลดไฟล์ไม่สำเร็จ")
merged_df
```
