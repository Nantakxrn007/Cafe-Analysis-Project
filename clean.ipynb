```python
from haversine import haversine, Unit

point_rangsit = (13.9904187, 100.5949663)  # รังสิต
point_khlong_rangsit = (13.9811737, 100.5983345)  # คลองรังสิต
point_lak_hok = (13.9659073, 100.5924792)  # หลักหก

def add_distance_columns(df):
    df['distance_rangsit'] = df.apply(lambda row: haversine((row['lat'], row['long']), point_rangsit, unit=Unit.KILOMETERS), axis=1)
    df['distance_khlong_rangsit'] = df.apply(lambda row: haversine((row['lat'], row['long']), point_khlong_rangsit, unit=Unit.KILOMETERS), axis=1)
    df['distance_lak_hok'] = df.apply(lambda row: haversine((row['lat'], row['long']), point_lak_hok, unit=Unit.KILOMETERS), axis=1)
    return df
```

```python
df_1 = add_distance_columns(df_1)
df_2 = add_distance_columns(df_2)
df_3 = add_distance_columns(df_3)
```

```python
df_3_ci = pd.read_csv('โซน คลองหนึ่ง\df_ci_combined.csv')
df_1_ci = pd.read_csv('โซน หลักหก\df_ci_combined11.csv')
df_2_ci = pd.read_csv('โซน รังสิต\df_ci_combined11.csv')
```

```python
df_1_ci['customer']
```

```python
import pandas as pd
import re
from datetime import datetime
import numpy as np
```

```python
def extract_and_convert(text):
    if pd.isna(text):
        return pd.Series([np.nan, np.nan])
    
    match = re.search(r'เวลา (\d{2}:\d{2}).*?ประมาณ (\d+)%', text)
    if match:
        # แปลงเวลาเป็น datetime.time
        try:
            time_val = datetime.strptime(match.group(1), "%H:%M").time()
        except ValueError:
            time_val = np.nan

        # แปลงเปอร์เซ็นต์เป็น int
        percent_val = int(match.group(2))

        return pd.Series([time_val, percent_val])
    else:
        return pd.Series([np.nan, np.nan])


```

```python
df_1_ci[['time', 'percent']] = df_1_ci['customer'].apply(extract_and_convert)
df_2_ci[['time', 'percent']] = df_2_ci['customer'].apply(extract_and_convert)
df_3_ci[['time', 'percent']] = df_3_ci['customer'].apply(extract_and_convert)
```

```python
from datetime import time

def float_to_time(x):
    if pd.isna(x):
        return None
    hour = int(x)
    minute = int((x - hour) * 100)  # ใช้ *100 ตามที่คุณแจ้ง
    return time(hour, minute)

```

```python
df_1_op['open_time'] = df_1_op['open'].apply(float_to_time)
df_1_op['open_close'] = df_1_op['close'].apply(float_to_time)
df_2_op['open_time'] = df_2_op['open'].apply(float_to_time)
df_2_op['open_close'] = df_2_op['close'].apply(float_to_time)
df_3_op['open_time'] = df_3_op['open'].apply(float_to_time)
df_3_op['open_close'] = df_3_op['close'].apply(float_to_time)
```
